```python
def print_3_times(value,n):
    for i in range(n):
        print(value)


```

```python
def print_n_times(n,*values):
    for i in range(n):
        for value in values:
            print(value)
        print()

```

```python
print_n_times(10,"hi")
```

```python
def print_k_times(*values,k=2):
    for i in range(k):
        for value in values:
            print(value)
        print()


print_k_times("hi",k=10)
```

```python
def sum_all(start,end):
    output=0

    for i in range(start,end+1):
        output+=i
    return output

```

```python
def factorial(n):
    if n ==0:
        return 1

    else:
        return n*factorial(n-1)


factorial(2)
```

```python
import random
hanguls = list("가나다라마바사")

with open("info.txt","w") as file:
    for i in range(1000):
        name = random.choice(hanguls)+random.choice(hanguls)

        weight = random.randrange(40,100)
        height = random.randrange(140,200)

        file.write("{},{},{}\n".format(name,weight,height))
```

```python
with open("info.txt","r") as file:
    for line in file :
        (name, weight, height) = line.strip().split(",")
        if (not name) or (not weight) or (not height):
            continue

        bmi = int(weight)/((int(height)/100)**2)
        #result=""
        if 25<=bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상 체중"
        else:
            result = "저체중"

        print('\n'.join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "BMI: {}",
            "결과: {}",
        ]).format(name, weight, height, bmi, result))
        print()
```

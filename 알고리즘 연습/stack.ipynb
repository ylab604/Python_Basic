```python
from collections import deque

queue = deque()
queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()
print(queue)
queue.reverse()
print(queue)
```

```python
def recursive_function(i):
    if i == 100:
        return
    print(i, "번째 재귀 함수에서", i+1,"번쨰 재귀 함수를 호출합니다.")
    recursive_function(i+1)

recursive_function(1)
```

```python
def factiorial_iterative(n):
    result =1

    for i in range(1,n+1):
        result*=i
    return result

factiorial_iterative(5)
```
